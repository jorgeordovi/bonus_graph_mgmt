```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
x = [0,1,2,3,4]
y= [0,1,4,9,16]
```

```python
plt.plot(x, y, color='#FF5733') # using hex code
plt.title('Color Hex Specicification')
plt.show()
```

```python
plt.plot(x, y, color='blue')
plt.title('Using Basic Strin Colors')
plt.show()

plt.plot(x, y, color=(0.1, 0.2, 0.5))
plt.title('Specifice RGB Setting')
plt.show()
```

```python
## Colormaps

z = np.linspace(0,1,100)
colores = plt.cm.viridis(z)

for i in range(len(z)):
    plt.plot(x, [v * (i*1) for v in y], color=colores[i])
plt.title('Use Colormaps')
plt.show()
```

```python
## Posible linestyles (-)(--)(-.)(:)

plt.plot(x, y, linestyle=':')
plt.title('Using Basic Linestyle')
plt.show() 
```

```python
## Adding Markers (o)(s)(^)

plt.plot(x,y, marker='^')
plt.title('Types of Merkers Avail')
plt.show()
```

```python
## Combine Markers and line types Said (--)(o)
## Line width 
plt.plot(x, y, color='orange', marker='o', linestyle='--', linewidth=2)
plt.title('Using Basic Strin Colors')
plt.show()
```

```python
## add grid and AX legend

plt.plot(x, y, color='orange', marker='o', linestyle='--', linewidth=2, label='Data')
plt.title('Using Advanced Personalized')
plt.xlabel('X Axis')
plt.ylabel('Y Label')
plt.legend()
plt.grid(True, color='white')
plt.show()
```
